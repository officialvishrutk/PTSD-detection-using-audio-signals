```python
import os
import pandas as pd
from fastai.vision.all import *

# Define the path to the directory containing the images
data_path = Path(r"G:\My Documents\project")

# Define the labels for all emotion folders
label_mapping = {
    'stress': 'stressed',
    'non_stress': 'non_stressed',
    'angry': 'stressed',
    'calm': 'non_stressed',
    'disgusted': 'stressed',
    'fearful': 'stressed',
    'happy': 'non_stressed',
    'neutral': 'non_stressed'
}

# Get a list of all image file paths and corresponding labels
file_paths = []
labels = []

# Loop through each emotion folder
for emotion_folder in os.listdir(data_path):
    # Skip non-directory entries
    if not os.path.isdir(data_path / emotion_folder):
        continue

    # Check if the folder is relevant
    label = label_mapping.get(emotion_folder, 'other_emotion')

    # Loop through each image file in the emotion folder
    for file_name in os.listdir(data_path / emotion_folder):
        # Skip non-file entries
        if not os.path.isfile(data_path / emotion_folder / file_name):
            continue

        # Append the file path and label to the lists
        file_paths.append(data_path / emotion_folder / file_name)
        labels.append(label)

# Create a DataFrame from the file paths and labels
df = pd.DataFrame({'file_paths': file_paths, 'labels': labels})
# Define data augmentation transformations
aug_transforms = aug_transforms(mult=2.0, do_flip=True, flip_vert=True, max_rotate=10.0, max_zoom=1.1, max_lighting=0.2, max_warp=0.2, p_affine=0.75, p_lighting=0.75)

# Create a DataBlock
dblock = DataBlock(blocks=(ImageBlock, CategoryBlock),
                   get_x=ColReader('file_paths'),
                   get_y=ColReader('labels'),
                   splitter=RandomSplitter(valid_pct=0.2),
                   item_tfms=Resize(460),
                   batch_tfms=[*aug_transforms, Normalize.from_stats(*imagenet_stats)])

# Create a DataLoader
dls = dblock.dataloaders(df, bs=32)

# Create a CNN model using a pretrained ResNet50 architecture
learn = cnn_learner(dls, resnet50, metrics=accuracy)

```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load the audio file
audio_path = r'C:\Users\V_KHAT\Downloads\2055 Rue Bishop 4.wav'
y, sr = librosa.load(audio_path, sr=None)  # sr=None to preserve the original sampling rate

# Generate the Mel spectrogram
mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=512, n_mels=128)

# Convert to decibels (log scale)
mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

# Display the Mel spectrogram
plt.figure(figsize=(10, 4))
librosa.display.specshow(mel_spectrogram_db, sr=sr, hop_length=512, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram')
plt.show()
plt.savefig('mel_spectrogram.jpg')
```

```python
import os
import pandas as pd
from fastai.vision.all import *

# Define the path to the directory containing the images
data_path = Path(r"G:\My Documents\project")

# Define the labels for all emotion folders
label_mapping = {
    'stress': 'stressed',
    'non_stress': 'non_stressed',
    'angry': 'stressed',
    'calm': 'non_stressed',
    'disgusted': 'stressed',
    'fearful': 'stressed',
    'happy': 'non_stressed',
    'neutral': 'non_stressed',
    'sad': 'stressed',
    'surprised': 'non_stressed'
}

# Get a list of all image file paths and corresponding labels
file_paths = []
labels = []

# Loop through each emotion folder
for emotion_folder in os.listdir(data_path):
    # Skip non-directory entries
    if not os.path.isdir(data_path / emotion_folder):
        continue

    # Check if the folder is relevant
    if emotion_folder not in label_mapping:
        print(f"Warning: {emotion_folder} not found in label_mapping")
        continue

    label = label_mapping[emotion_folder]

    # Loop through each image file in the emotion folder
    for file_name in os.listdir(data_path / emotion_folder):
        # Skip non-file entries
        if not os.path.isfile(data_path / emotion_folder / file_name):
            continue

        # Append the file path and label to the lists
        file_paths.append(data_path / emotion_folder / file_name)
        labels.append(label)

# Create a DataFrame from the file paths and labels
df = pd.DataFrame({'file_paths': file_paths, 'labels': labels})

# Print out the distribution of labels
print(df['labels'].value_counts())

# Define data augmentation transformations
aug_transforms = aug_transforms(mult=2.0, do_flip=True, flip_vert=True, max_rotate=10.0, max_zoom=1.1, max_lighting=0.2, max_warp=0.2, p_affine=0.75, p_lighting=0.75)

# Create a DataBlock
dblock = DataBlock(blocks=(ImageBlock, CategoryBlock),
                   get_x=ColReader('file_paths'),
                   get_y=ColReader('labels'),
                   splitter=RandomSplitter(valid_pct=0.2),
                   item_tfms=Resize(460),
                   batch_tfms=[*aug_transforms, Normalize.from_stats(*imagenet_stats)])

# Create a DataLoader
dls = dblock.dataloaders(df, bs=32)

# Create a CNN model using a pretrained ResNet50 architecture
learn = cnn_learner(dls, resnet50, metrics=accuracy)
```
